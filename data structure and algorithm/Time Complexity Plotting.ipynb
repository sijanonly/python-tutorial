### Let's observe the complexity of different approach of creating lists
<meta property="og:image" content=""/>

```python

def append(n):
    # append
    l = []
    for i in range(n):
        l.append(i)

def concatenation(n):
    # concatenation
    l = []
    for i in range(n):
        l = l + [i]

def list_comprehension(n):
    # list comprehension
    l = [i for i in range(n)]

def list_constructor(n):
    # range + list constructor
    l = list(range(n))


```

```python
%matplotlib inline
from matplotlib import pyplot

import timeit
from functools import partial
def plot_time_complexity(fn, min_value, max_value, incr, no_of_tests):
    """
    Plotting time complexity based on given function.
    """
    x = []
    y = []
    for i in range(min_value, max_value, incr):
        N = i
        timer = timeit.Timer(partial(fn, N))
        t = timer.timeit(number=no_of_tests)
        x.append(i)
        y.append(t)
    p1 = pyplot.plot(x, y, 'o', label = fn.__name__)
    pyplot.legend(loc='upper left')
    pyplot.xlabel('data size', fontsize=18)
    pyplot.ylabel('execution time', fontsize=16)
```

```python

plot_time_complexity(concatenation, 1, 100, 1, 1000)
plot_time_complexity(append, 1, 100, 1, 1000)
plot_time_complexity(list_comprehension, 1, 100, 1, 1000)
plot_time_complexity(list_constructor, 1, 100, 1, 1000)
```
